# Monte Carlo Technique
### An experimental technique used to calculate the value of pi.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_graph(x, y, x_, y_, pi):
    fig=plt.figure(figsize=(4, 4))
    ax=fig.add_axes([0,0,1,1])
    ax.scatter(x, y, color='b')
    ax.scatter(x_, y_, color='r')
    ax.set_xlabel('Grades Range')
    ax.set_ylabel('Grades Scored')
    ax.set_title(f'Pi value - {pi}')
    plt.show()
```

```python
def calculte_pi(r, no_of_beans, times):
    pi_value_list = []
    for i in range(0, times):
        x = np.random.default_rng().uniform(-r, r, no_of_beans)
        y = np.random.default_rng().uniform(-r, r, no_of_beans)
        beans_in_circle = 0
        x_in = []
        x_out = []
        y_in = []
        y_out = []
        for j in range(0, no_of_beans):
            bean_x = x[j]
            bean_y = y[j]
            if bean_x**2+bean_y**2<=r**2:
                beans_in_circle = beans_in_circle + 1
                x_in.append(bean_x)
                y_in.append(bean_y)
            else:
                x_out.append(bean_x)
                y_out.append(bean_y)
        pi = 4*(beans_in_circle/no_of_beans)
        plot_graph(x_in, y_in, x_out, y_out, pi)
#         print("exp - ",4*(beans_in_circle/no_of_beans))
        pi_value_list.append(pi)
    
    return np.mean(pi_value_list)
```

```python
calculte_pi(1, 10000, 5)
```
